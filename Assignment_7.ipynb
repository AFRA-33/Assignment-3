##### Make a class called Employee. The __init__() method for Employee should store its different attributes: name, number, address, salary, post, year of joining, education, and his year of experience.

```python
# print("Please enter your details.......")
# name = input("What is your name? ")
# number = int(input("Your Phone number?"))
# address = input("Your Address?")
# salary = int(input("Your Salary?"))
# post = input("Your post?")
# year_of_joining = int(input("Your year_of_joining?"))
# education = input("Your education?")
# years_of_experience = int(input("Your years_of_experience?"))


class Employee:
    def __init__(self, name, number, address, salary, post, year_of_joining, education, years_of_experience):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.years_of_experience = years_of_experience

    def describe_employee(self):
        print()
        #print("==================================== YOUR DETAILS ARE ========================================================")
        return (f"Employee Name: {self.name}\n"
                f"Employee Number: {self.number}\n"
                f"Address: {self.address}\n"
                f"Salary: {self.salary}\n"
                f"Post: {self.post}\n"
                f"Year of Joining: {self.year_of_joining}\n"
                f"Education: {self.education}\n"
                f"Years of Experience: {self.years_of_experience}")

    def update_year_of_experience(self, update_year_of_experience): 
        print("===========================Updates the employee's years of experience =============================")
        self.years_of_experience = update_year_of_experience 
        print(f"Years of experience updated to {self.years_of_experience}")

    def increase_salary(self,increase_salary):
        print("======================================== Increases the employee's salary by a specified amount ============================" )
        self.salary += increase_salary 
        print(f"Salary increased by {increase_salary}. New salary is {self.salary}")



# employee = Employee(name, number, address, salary, post, year_of_joining, education,years_of_experience)
employe = Employee(
    name="John Doe",
    number=12345,
    address="123 Main St, Cityville",
    salary=75000,
    post="Software Engineer",
    year_of_joining=2015,
    education="Bachelor's in Computer Science",
    years_of_experience=8
)

print(employe.describe_employee())
print(employe.update_year_of_experience(2008))
print(employe.increase_salary(2008))
```
